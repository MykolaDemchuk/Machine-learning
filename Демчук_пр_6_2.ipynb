```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
```

```python
#Завантажити набір даних
breast_cancer = load_breast_cancer()
a = breast_cancer.data
b = breast_cancer.target

a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=42)
```

```python
#Розділити дані на тренувальний та тестовий набори
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=42)
```

```python
#Застосувати випадковий ліс для визначення важливих ознак.
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(a_train, b_train)

vaz_oznaka= rf_classifier.feature_importances_
```

```python
#Побудувати графік або таблицю, щоб показати важливість кожної ознаки в аналізі
def plot_vaz_oznaka(features, feature_importance):
    plt.figure(figsize=(10, 6))
    plt.barh(features, feature_importance)
    plt.xlabel('Важливість ознак')
    plt.title('Важливість ознак у моделі Random Forest')
    plt.show()

features = breast_cancer.feature_names
plot_vaz_oznaka(features, vaz_oznaka)
```

```python
#За допомогою цієї таблиці ми дізналися, що "worst area","worst convcave points" більше всього впливають на вихідні дані.
```

```python
#Обчислити точність для двох алгоритмів і порівняти результати
svm_classifier = SVC(random_state=42)
svm_classifier.fit(a_train, b_train)
svm_accuracy = svm_classifier.score(a_test, b_test)

rf_accuracy = rf_classifier.score(a_test, b_test)

print(f"SVM точність: {svm_accuracy}")
print(f"Random Forest точність: {rf_accuracy}")
```
