```python
#Прочитайте CSV файл (використовуйте метод read_csv):
globa = pd.read_csv('/content/sample_data/Global YouTube Statistics.csv', encoding='latin1')
globa
```

```python
#Виведіть перші п'ять рядків (використовується метод head)
glob_1 = globa.head(5)
print(glob_1)
```

```python
#Виведіть розміри датасету (використовуйте метод shape)
print(glob_1.shape)
```

```python
#Перевірте, чи у всіх рядків вистачає даних: виведіть кількість пропусків  у кожному зі стовпців (використовуйте методи isna та sum). Якщо є пропуски, замінить їх на NaN, потім тип даних на числовий.
glob_1 = glob_1.fillna('Nan')
propusk = glob_1.isna().sum()
print(glob_1)
print(propusk)
#Тут не все рішення, тому що я заплутався в значеннях та далі таблиця ломалася, після змін.
```

```python
#Замініть комірки з пропущеними значенями на середні за стовпцем. І тип даних на float
glob_1 = glob_1.fillna(glob_1.mean())
propusk = glob_1.isna().sum()
print(glob_1)
print(propusk)
```

```python
#Перевірте, скільки унікальних країн представлено у колонці "Country"
stranu = glob_1['Country'].nunique()
print(stranu)
```

```python
#Побудуйте діаграму розподілу переглядів (використовуйте kind='hist')
glob_1['video views'].plot(kind='hist', title='Розподіл переглядів')
```

```python
#Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean)
max = glob_1['video views'].max()
min = glob_1['video views'].min()
mean = glob_1['video views'].mean()
print(max)
print(min)
print(mean)
```

```python
#найдіть країну, де найбільше відео було завантажено на YouTube
zavan = glob_1['uploads'].idxmax()
zavann = glob_1.loc[zavan, 'Country']
print(zavann)
```

```python
#Знайдіть відео (виведіть його назву) з найбільшою та найменшою кількістю завантажень uploads.
max_up = glob_1[glob_1['uploads'] == glob_1['uploads'].max()]['Title'].values[0]
min_up = glob_1[glob_1['uploads'] == glob_1['uploads'].min()]['Title'].values[0]
print(max_up)
print(min_up)
```
